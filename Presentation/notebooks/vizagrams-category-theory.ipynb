## Spinoff - Category Theory

Creating a meta-language to draw diagrams for Category Theory.

```julia
using Pkg
Pkg.activate(".")

using Vizagrams
using LinearAlgebra
using LaTeXStrings
# using Graphs
# using StructArrays
# using DataFrames
# using Random
# using NetworkLayout

include("catdiagrams.jl");
include("vizdiagram.jl");
```

```julia
A = obj(:A)
d = A + idmor(A)

savefigure(d,filename="1Cat.pdf");
vizcat(d)
```

```julia
A = obj(:A)
B = obj(:B,x=1)
C = obj(:C,x=2)
d = A + B + mor(A,B,"f") + idmor(A) + idmor(B)

savefigure(d,filename="2Cat.pdf");
vizcat(d)
```

```julia
d = A + idmor(A)+
    idmor(B) + B + mor(A,B,"f") +
    idmor(C) + C + mor(B,C,"g") +
    T(0,-0.2)mor(A,C,L"g\circ f",labeltransform=T(0,-0.2),curve=-0.5)

savefigure(d,filename="3Cat.pdf");
vizcat(d)
```

```julia
d = A +
    B + mor(A,B,"f") +
    C + mor(B,C,"g")

savefigure(d,filename="3Catsimple.pdf");
vizcat(d)
```

```julia
o1 = obj(:1)
o2 = obj(:2,x=1)
dots = obj("...",x=2)
on = obj(:N,x=3)
d = o1 + idmor(o1) + o2 + idmor(o2) + dots + on + idmor(on)

savefigure(d,filename="DiscreteCategory.pdf");
vizcat(d)
```

```julia
include("catdiagrams.jl");
A = obj(:A)
FA = obj(:FA,x=1)
idF = A + FA +mor(A,B,level=2, length=0.8,label=L"F") + idmor(A) +
    idmor(FA,label=L"F(id_A) = id_{FA}")

vizcat(idF)
```

```julia
Cat = T(0,0.1)TextMark(text=L"\scrC",fontsize=0.2,anchor=:e)
Catop = T(0,0.1)TextMark(text=L"\scrC^{op}",fontsize=0.2,anchor=:e)

A = obj(:A)
B = obj(:B,x=1,y=1.0)
C = obj(:C,x=1)
D = obj(:D,x=1,y=-1.0)
d = A + B + mor(A,B) + C + mor(A,C) + D + mor(A,D)

rect = T(0.6,0)S(:fillOpacity=>0.0,:stroke=>:black)*Rectangle(h=3,w=2)
d = d + T(0.5,1.5)Cat + rect



dop = A + B + mor(B,A) + C + mor(C,A) + D + mor(D,A)
dop = dop + T(0.5,1.5)Catop + rect


d = d + T(2.5,0)dop

savefigure(d,filename="DualCategory.pdf");
vizcat(d)
```

```julia
A = obj(:A)
B = obj(:B,x=1)
C = obj(:C,x=1,y=-1)

c = A + B + C
c = c + mor(A,B,"f") + mor(B,C,"g") + mor(A,C, L"g\circ f", labeltransform=T(-0.3,-0.2))
vizcat(c)
```

```julia
include("catdiagrams.jl");
FA = obj(:FA)
FB = obj(:FB,x=1)
FC = obj(:FC,x=1,y=-1)
Fd = FA + FB + FC

fd = Fd + mor(FA,FB,"Ff") + mor(FB,FC,"Fg") + mor(FA,FC, L"Fg\circ Ff", labeltransform=T(-0.4,-0.2))
fd = c + T(2.7,0)*fd + T(1.5,-.5)Morphism(level=2,label=L"F",length=0.8, fontsize=0.12,labeltransform=T(0,0.03))

contrafd = Fd + mor(FB,FA,"Ff") + mor(FC,FB,"Fg") + mor(FC,FA, L"Ff \ \circ Fg", labeltransform=T(-0.1,-0.2))
contrafd = c + T(2.8,0)*contrafd + T(1.5,-.5)Morphism(level=2,label=L"F",length=0.8, fontsize=0.12,labeltransform=T(0,0.03))

functordiagram = T(-0.1,-0.5)idF + T(2.4,0)*fd + T(-0.3,0.4)TextMark(text=L"\text{Covariant Functor}",fontsize=0.1,anchor=:e)
contradiagram  = T(-0.1,-0.5)idF + T(2.4,0)*contrafd + T(-0.3,0.4)TextMark(text=L"\text{Contravariant Functor}",fontsize=0.1,anchor=:e)
functordiagram = T(0,-2)*(contradiagram) + functordiagram

savefigure(functordiagram,filename="Functor.pdf");
vizcat(functordiagram)
```

```julia
FA = obj(:FA)
GA = obj(:GA,x=1)
FB = obj(:FB,x=0,y=-1)
GB = obj(:GB,x=1,y=-1)

d = FA + GA + FB + GB
d = d +
    mor(FA,GA,L"\alpha_A") +
    mor(FA,FB,L"Ff",labeltransform=T(-0.3,0)) +
    mor(FB,GB,L"\alpha_B",labeltransform=T(0,-0.2)) +
    mor(GA,GB,L"G f")


t = TextMark(text=L"\alpha_B \circ Ff  \ = Gf \ \circ \alpha_A",fontsize=0.12,anchor=:e)
t = TextMark(text=L"Gf \ \circ \alpha_A = \alpha_B \circ Ff",fontsize=0.12,anchor=:e)
d = d + T(0,-1.8)acenter(d,t)*t

savefigure(d,filename="NaturalTransformation.pdf");
vizcat(d)
```

```julia
include("catdiagrams.jl");
C = obj(L"\scrC")
D = obj(L"\scrD",x=1)

d = C + D
d = d + 
    T(0,0.3)mor(C,D,"F",curve=0.3) +
    T(0,-0.3)mor(C,D,"H",curve=-0.3,labeltransform=T(0,-0.2)) +
    mor(C,D,"G",labeltransform=T(0,-0.10))+
    T(1.2/2-0.02,0.35)Morphism(pos=[[0,0],[0,-0.2]],level=2,label=L"\ \alpha") +
    T(1.2/2-0.02,-0.35)Morphism(pos=[[0,0.2],[0,0.]],level=2,label=L"\ \beta")

savefigure(d,filename="VerticalComposition.pdf");
vizcat(d)
```

```julia
include("catdiagrams.jl");
C = obj(L"\scrC")
D = obj(L"\scrD",x=1)
F = obj(L"\scrF",x=2)

d = C + D + F
d = d + 
    T(0,0.2)mor(C,D,"F",curve=0.2) +
    T(0,-0.2)mor(C,D,"G",curve=-0.2,labeltransform=T(0,-0.2)) +
    T(1.2/2-0.02,0.)Morphism(pos=[[0,0.2],[0,-0.2]],level=2,label=L"\ \alpha")+
    T(0,0.2)mor(D,F,"H",curve=0.2) +
    T(0,-0.2)mor(D,F,"K",curve=-0.2,labeltransform=T(0,-0.2)) +
    T(1.2/2+1.2-0.02,0.)Morphism(pos=[[0,0.2],[0,-0.2]],level=2,label=L"\ \beta")

savefigure(d,filename="HorizontalComposition.pdf");
vizcat(d)
```

```julia
N = obj(L"\bbN")
d = T(0.02,0)N + mor(N,L"id_ℕ = 0") + mor(N,curve=1,label=L"1") + mor(N,curve=2,label=L"2") +  mor(N,curve=3,label=L"\vdots")

savefigure(d,filename="NMonoid.pdf");
vizcat(d)
```

```julia
MMM = obj(L"M \times M \times M")
MM1 = obj(L"M\times M",x=1.5)
MM2 = obj(L"M\times M",x=0,y=-1)
Mi = obj(:M,x=1.5,y=-1)

d = MMM + MM1 + MM2 + Mi
d1 = d +
    mor(MMM,MM1,L"id_M \times \mu") +
    mor(MMM,MM2,L"\mu \times id_M",labeltransform=T(-0.55,0)) +
    mor(MM1,Mi,L"\mu",labeltransform=T(0,0)) +
    mor(MM2,Mi,L"\mu")

MM = obj(L"M\times M",x=1.)
M1 = obj(:M)
M2 = obj(:M,x=2)
M3 = obj(:M,x=1.,y=-1)
d2 = MM + M1 + M2 + M3
d2 = d2 +
    mor(M1,MM,L"\eta \times id_M")+
    mor(M2,MM,L"id_M \times \eta")+
    mor(MM,M3,L"\mu") +
    mor(M1,M3,L"id_M",labeltransform=T(-0.3,-0.1)) +
    mor(M2,M3,L"id_M",labeltransform=T( 0.1,-0.1))

d = d1 + T(3,0)*d2

savefigure(d,filename="MonoidalCategory.svg");
savefigure(d,filename="../../presentation/figs/MonoidalCategory.svg");
vizcat(d)
```

```julia
TTT = obj(L"T \circ T \circ T")
TT1 = obj(L"T\circ T",x=1.5)
TT2 = obj(L"T\circ T",x=0,y=-1)
Ti = obj(:T,x=1.5,y=-1)

d = TTT + TT1 + TT2 + Ti
d1 = d +
    mor(TTT,TT1,L"id_T \ast \mu") +
    mor(TTT,TT2,L"\mu \ast id_T",labeltransform=T(-0.5,0)) +
    mor(TT1,Ti,L"\mu",labeltransform=T(0,0)) +
    mor(TT2,Ti,L"\mu")

TT = obj(L"T\circ T",x=1.)
T1 = obj(:T)
T2 = obj(:T,x=2)
T3 = obj(:T,x=1.,y=-1)
d2 = TT + T1 + T2 + T3
d2 = d2 +
    mor(T1,TT,L"\eta \ast id_T")+
    mor(T2,TT,L"id_T \ast \eta")+
    mor(TT,T3,L"\mu") +
    mor(T1,T3,L"id_T",labeltransform=T(-0.3,-0.1)) +
    mor(T2,T3,L"id_T",labeltransform=T( 0.1,-0.1))

d = d1 + T(3,0)*d2

savefigure(d,filename="Monad.pdf");
savefigure(d,filename="../../presentation/figs/Monad.svg");
vizcat(d)
```

```julia
A = obj(L"A",x=-1.)
B = obj(L"B",x=1.)
C = obj(L"S",x=0.,y=-1)
d = B + A + C +
    mor(A,B,L"\phi")+
    mor(A,C,L"f",labeltransform=T(-0.3,-0.1)) +
    mor(B,C,L"g",labeltransform=T( 0.1,-0.1))

cat = TextMark(text=L"\scrC",fontsize=0.15)
d =  d + T(0,0.6)*cat

d = TextMark(text=L"\cong", fontsize=0.2) → (T(0.5,0),d)

Af = obj("(A, \\ f \\ )",x=-1.)
Bg = obj("(B,\\ g \\ )",x=1.)

slice = Af + Bg + mor(Af,Bg,L"\phi_{A,B}") + T(0,0.6)TextMark(text=L"\scrC / S",fontsize=0.15)
d =  slice → (T(0.5,0),d)


savefigure(d,filename="../../presentation/figs/SliceCategory.pdf");
savefigure(d,filename="../../presentation/figs/SliceCategory.svg");

vizcat(d)
```

## F-Algebra

```julia
FA = obj(:FA)
FB = obj(:FB,x=1)
A = obj(:A,x=0,y=-1)
B = obj(:B,x=1,y=-1)

d = FA + A + FB + B
d = d +
    mor(FA,A,L"\phi",labeltransform=T(-0.2,0)) +
    mor(FA,FB,L"Ff") +
    mor(FB,B,L"\psi") +
    mor(A,B,L"f",labeltransform=T(0,-0.2))


t = TextMark(text=L"\psi \ \circ Ff \ = f \ \circ \phi",fontsize=0.12,anchor=:e)
d = d + T(0,-1.8)acenter(d,t)*t

savefigure(d,filename="F-algebra.pdf");
vizcat(d)
```

```julia
FA = obj(:FI)
FB = obj(:FA,x=1)
A = obj(:I,x=0,y=-1)
B = obj(:A,x=1,y=-1)

d = FA + A + FB + B
d = d +
    mor(FA,A,L"in",labeltransform=T(-0.3,0)) +
    mor(FA,FB,L"F cata_\phi") +
    mor(FB,B,L"\phi") +
    mor(A,B,L"cata_\phi",labeltransform=T(0,-0.2))


t = TextMark(text=L"\phi \ \circ F cata_\phi \ = cata_\phi \ \circ \ in",fontsize=0.12,anchor=:e)
d1 = d + T(0,-1.8)acenter(d,t)*t

FA = obj(:FI)
FB = obj(:FA,x=1)
A = obj(:I,x=0,y=-1)
B = obj(:A,x=1,y=-1)

d = FA + A + FB + B
d = d +
    mor(A,FA,L"in^{-1}") +
    mor(FA,FB,L"F cata_\phi") +
    mor(FB,B,L"\phi") +
    mor(A,B,L"cata_\phi",labeltransform=T(0,-0.2))


t = TextMark(text=L"cata_\phi = \phi \circ F cata_\phi \circ in^{-1} ",fontsize=0.12,anchor=:e)
d2 = d + T(0,-1.8)acenter(d,t)*t

d = d1 → (T(1,0),d2)

savefigure(d,filename="catamorphism.pdf");
vizcat(d)
```
