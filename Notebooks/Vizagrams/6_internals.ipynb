# Vizagrams Internals

```julia
using Pkg
Pkg.activate("../.")
using Vizagrams
```

```julia
struct Body <: Mark end

function Vizagrams.ζ(b::Body)
    Rectangle(h=2,w=0.2) +
    Line([[0,1],[-1.,-0.3]]) +
    Line([[0,1],[1.,-0.3]]) +
    Line([[-0.1,-1],[-0.4,-2.5]]) +
    Line([[0.1,-1],[0.4,-2.5]]) 
end

struct Person <: Mark end
function Vizagrams.ζ(p::Person)
    Face() + T(0,-2)Body()
end
```

```julia
d = Person()
draw(d)
```

```julia
ζ(d) |> treediagram |> draw
```
